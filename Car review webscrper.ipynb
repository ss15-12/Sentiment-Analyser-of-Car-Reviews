```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# function to print sentiments
# of the sentence.
def sentiment_scores(sentence):

	# Create a SentimentIntensityAnalyzer object.
	sid_obj = SentimentIntensityAnalyzer()

	# polarity_scores method of SentimentIntensityAnalyzer
	# object gives a sentiment dictionary.
	# which contains pos, neg, neu, and compound scores.
	sentiment_dict = sid_obj.polarity_scores(sentence)
	
# 	print("Overall sentiment dictionary is : ", sentiment_dict)
	print("sentence was rated as ", sentiment_dict['neg']*100, "% Negative")
	print("sentence was rated as ", sentiment_dict['neu']*100, "% Neutral")
	print("sentence was rated as ", sentiment_dict['pos']*100, "% Positive")
	print('')
# 	print("Sentence Overall Rated As", end = " ")
	# decide sentiment as positive, negative and neutral
#	if sentiment_dict['compound'] >= 0.05 :
#		print("Positive")

#	elif sentiment_dict['compound'] <= - 0.05 :
#		print(sentence)
#		print("Negative")

#	else :
#		print(sentence)        
#		print("Neutral")
for i in concatenated['Review']:
    sentence=i
    sentiment_scores(sentence)
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt


# Creating dataset
cars = ['Negative', 'Positive', 'Neutral']

data= np.array([1.2233333333333334,
31.93777777777778,
66.84055555555557])


# Creating explode data
explode = (0.1, 0.1, 0.1)

# Creating color parameters
colors = ( "Red", "Green", "yellow")

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }

# Creating autocpt arguments
def func(pct, allvalues):
	absolute = int(pct / 100.*np.sum(allvalues))
	return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
								autopct = lambda pct: func(pct, data),
								explode = explode,
								labels = cars,
								shadow = True,
								colors = colors,
								startangle = 90,
								wedgeprops = wp,
								textprops = dict(color ="maroon"))

# Adding legend
ax.legend(wedges, cars,
		title ="Cars",
		loc ="center left",
		bbox_to_anchor =(1, 0, 0.5, 1))

plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Sentiment Percentage Distribution For Mahindra Scorpio")

# show plot
plt.show()
```

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import opinion_lexicon
import re
df = concatenated
lemma = WordNetLemmatizer()
stop_words = stopwords.words('english')
def text_prep(x: str) -> list:
     corp = str(x).lower() 
     corp = re.sub('[^a-zA-Z]+',' ', corp).strip() 
     tokens = word_tokenize(corp)
     words = [t for t in tokens if t not in stop_words]
     lemmatize = [lemma.lemmatize(w) for w in words]
     return lemmatize
preprocess_tag = [text_prep(i) for i in df['Review']]
df["preprocess_txt"] = preprocess_tag
df['total_len'] = df['preprocess_txt'].map(lambda x: len(x))
#file = open('negative-words.txt', 'r')
neg_words = opinion_lexicon.negative()
#file = open('positive-words.txt', 'r')
pos_words = opinion_lexicon.positive()
num_pos = df['preprocess_txt'].map(lambda x: len([i for i in x if i in pos_words]))
```
