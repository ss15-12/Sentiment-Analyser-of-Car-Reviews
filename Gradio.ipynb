```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download("vader_lexicon")
sid = SentimentIntensityAnalyzer()

def sentiment_analysis(text):
    scores = sid.polarity_scores(text)
    del scores["compound"]
    return scores

sen = gr.Interface(
    fn=sentiment_analysis,
    inputs=gr.Textbox(placeholder="Enter a positive or negative sentence here..."),
    outputs="label",
    interpretation="default",
    examples=[["This is wonderful!"]])
```

```python
def ranking(text):
  if text=="XUV":
    scores={'performance':1,'steering':1,'comfort':0.4}
  elif text=="SCORPRIO":
    scores={'performance':1,'steering':1,'comfort':0.4}
  elif text=="THAR":
    scores={'performance':1,'steering':1,'comfort':0.4}
  elif text=="BOLERO":
    scores={'performance':1,'steering':1,'comfort':0.4}
  return scores
inputs = [ gr.Dropdown(["XUV", "SCORPIO", "THAR","BOLERO"], label="Which car?")]
demo1 = gr.Interface(
    fn=ranking,
    inputs=inputs,
    outputs="label")
```
